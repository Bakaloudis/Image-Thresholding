```python
import sys
import numpy as np
import matplotlib.pyplot as plt 

def main(threshold):

    from PIL import Image

    image = np.array(Image.open('trikoupi6.png')) # opening image as an array
    th = threshold

    #print(image)

    if(len(image.shape)<3): # check if image is rgb or gray scale
        grayImage = image
    elif len(image.shape)==3:
        grayImage = np.dot(image[...,:3], [0.2989, 0.5870, 0.1140]) # rbg to grayscale

    grayImage_bool = ( grayImage > float(th) ) * 255 # mult. by 255 to make True == White and False == Black 

    return grayImage_bool

    #Image.fromarray(np.uint8(grayImage_bool)).save(sys.argv[2]) # saving the image to disk

image = main(20)
print("threshold = " + str(20))
plt.imshow(image, cmap="gray")  # Showing the image
plt.show()

image = main(65)
print("threshold = " + str(65))
plt.imshow(image, cmap="gray")  # Showing the image
plt.show()

image = main(100)
print("threshold = " + str(100))
plt.imshow(image, cmap="gray")  # Showing the image
plt.show()

image = main(120)
print("threshold = " + str(120))
plt.imshow(image, cmap="gray")  # Showing the image
plt.show()

image = main(140)
print("threshold = " + str(140))
plt.imshow(image, cmap="gray")  # Showing the image
plt.show()
```
